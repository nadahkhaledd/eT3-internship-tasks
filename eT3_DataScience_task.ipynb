```python
# import nessecary libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prepare data
df = pd.read_csv('drinkMenu.csv')
df.drop_duplicates(inplace=True)
df.fillna(0.0, inplace=True)
```

```python
#print(df.head)
```

```python
#   drop uncessary columns

# df.drop(["Saturated Fat (g)", "Vitamin C (% DV)", "Vitamin A (% DV)", 
#          " Calcium (% DV)", "Iron (% DV)", " Dietary Fibre (g)", 
#          " Sodium (mg)", "Cholesterol (mg)", "Trans Fat (g) "], axis = 1, inplace=True)

df.drop(["Saturated Fat (g)", "Vitamin C (% DV)", 
        " Dietary Fibre (g)", " Sodium (mg)", "Cholesterol (mg)", "Trans Fat (g) "], axis = 1, inplace=True)
#print(df.head)
```

```python
# drink with the highest calories 
highestCaloriesDrink = df.loc[df['Calories'].idxmax()]

#plotting only the drink with hieghst calories
plt.title('Highest Calories drink')
plt.scatter(highestCaloriesDrink['Beverage'], highestCaloriesDrink['Calories'], color='blue')
plt.xlabel('Beverage', color = 'blue', fontsize = 15)
plt.ylabel('Calories', color = 'blue', fontsize = 15)
plt.show()

# top 5 drinks with highest calories
top5Cal = df.nlargest(5,"Calories")

#plotting top 5 drinks with highest calories
plt.figure(figsize=(8,5))
plt.tight_layout()
plt.title('Top 5 drinks with highest caloris')
plt.scatter(top5Cal['Beverage'], top5Cal['Calories'], color='red')
ax = plt.subplot()
plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
plt.xlabel('Beverage', color = 'r', fontsize = 15)
plt.ylabel('Calories', color = 'r', fontsize = 15)
plt.show()
```

```python
# drink with the highest sugars 
highestSugarsDrink = df.loc[df[' Sugars (g)'].idxmax()]

#plotting only drink with highest sugar
plt.title('Highest Sugars drink')
plt.scatter(highestSugarsDrink['Beverage'], highestSugarsDrink['Calories'], color='blue')
plt.xlabel('Beverage', color = 'blue', fontsize = 15)
plt.ylabel('Sugars (g)', color = 'blue', fontsize = 15)
plt.show()

# top 5 drinks with high sugars
top5Sug = df.nlargest(5,' Sugars (g)')

# plotting top 5 drinks with high sugars
plt.figure(figsize=(8,5))
plt.tight_layout()
plt.title('Top 5 drinks with high Sugars')
plt.scatter(top5Sug['Beverage'], top5Sug[' Sugars (g)'], color='red')
ax = plt.subplot()
plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
plt.xlabel('Beverage', color = 'r', fontsize = 15)
plt.ylabel(' Sugars (g)', color = 'r', fontsize = 15)
plt.show()
```
